# 第 1 章 使用函数绘制 matplotlib 的图表组成元素

## 1.1 绘制 matplotlib 图表组成元素的主要函数
在一个图形输出窗口中，底层是一个 Figure 实例（通常称为画布），保护一些可见和不可见的元素。

## 1.2 准备数据
导入第三方包 Numpy 和快速绘图模块 pyplot，其中科学计算包 Numpy 是 matplotlib 的基础。

```python
import matplotlib.pyplot as plt
import numpy as np

# linspace 表示在 -3.2 到 3.2 之间均匀地取 100 个数
x = np.linspace(-3.2, 3.2, 100)
y = np.sin(x)
y1 = np.random.randn(100)

# x：x 轴上地数值，y：y 轴上地数值
# ls：折线图地线条风格，supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
# lw：折线图地线条宽度，label：标记图形内容地标签文本
plt.plot(x, y, ls="-.", lw=2, label="plot figure")
plt.legend()
plt.show()
```

## 1.3 绘制 matplotlib 图表组成元素的函数用法

### 1.3.1 函数 plot()——展现变量的趋势变化

```python
import matplotlib.pyplot as plt
import numpy as np

# linspace 表示在 -3.2 到 3.2 之间均匀地取 100 个数
x = np.linspace(-3.2, 3.2, 100)
y = np.cos(x)
y1 = np.random.randn(100)

# x：x 轴上地数值，y：y 轴上地数值
# ls：折线图地线条风格，supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
# lw：折线图地线条宽度，label：标记图形内容地标签文本
plt.plot(x, y, ls="-.", lw=2, label="plot figure")
plt.legend()
plt.show()
```

### 1.3.2 函数 scatter()——寻找变量之间地关系

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.05, 10, 1000)
y = np.random.rand(1000)

# plt.scatter(x,y,c="b",lable="scatter figure")
# c：散点图中的标记的颜色
plt.scatter(x, y, c="b", label="scatter figure")
plt.legend()
plt.show()
```

### 1.3.3 函数 xlim()——设置 x 轴的数值显示范围

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.05, 10, 1000)
y = np.random.rand(1000)

plt.scatter(x, y, label="scatter figure")
plt.legend()

# plt.xlim(xmin, xmax)：设置 x 轴的数值显示范围
plt.xlim(0.05, 10)
plt.ylim(0, 1)

plt.show()
```

### 1.3.4 函数 xlabel()——设置 x 轴的标签文本

```python
import matplotlib.pyplot as plt
import numpy as np

# linspace 表示在 -3.2 到 3.2 之间均匀地取 100 个数
x = np.linspace(-3.2, 3.2, 1000)
y = np.sin(x)
y1 = np.random.randn(100)

plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend()

# plt.xlabel(string) 设置 x 轴的标签文本
plt.xlabel("x-axios")
plt.ylabel('y-axios')
plt.show()
```

### 1.3.5 函数 grid()——绘制刻度线的网格线

```python
import matplotlib.pyplot as plt
import numpy as np

# linspace 表示在 -3.2 到 3.2 之间均匀地取 100 个数
x = np.linspace(-3.2, 3.2, 100)
y = np.sin(x)
y1 = np.random.randn(100)


plt.plot(x, y, ls="-.", lw=2, label="plot figure")
plt.legend()

# 绘制刻度线的网格线
plt.grid(linestyle=":", color="r")
plt.show()
```

### 1.3.6 函数 axhline()——绘制平行于 x 轴的水平参考线

```python
import matplotlib.pyplot as plt
import numpy as np

# linspace 表示在 -3.2 到 3.2 之间均匀地取 100 个数
x = np.linspace(-3.2, 3.2, 100)
y = np.sin(x)
y1 = np.random.randn(100)

plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend()

plt.axhline(y=0.0, c="r", ls="--", lw=2)
plt.axvline(x=0.0, c="r", ls="--", lw=2)
plt.show()
```

### 1.3.7 函数 axvspan()——绘制垂直于 x 轴的参考区域

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3.2, 3.2, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend()

# plt.axvspan(xmin: 参考区域的起始位置, xmax：参考区域的终止位置, facecolor：参考区域的填充颜色, alpha：填充颜色的透明度)
plt.axhspan(ymin=0.0, ymax=0.5, facecolor="y", alpha=0.3)
plt.axvspan(xmin=0.5, xmax=1, facecolor="y", alpha=0.3)
plt.show()
```

### 1.3.8 函数 annotate()——添加图形内容细节的指向型注释文本

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3.2, 3.2, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend()

# plt.annotate(string：图形内容的注释文本, xy：被注释图形内容的位置坐标, 
# xytext：注释文本的位置坐标, weight：注释文本的位置坐标, 
# weight：注释文本的字体粗细风格, color：注释文本的字体颜色, 
# arrowprops：指示被注释内容的箭头的属性字典)
plt.annotate("maximum",
             xy=(np.pi/2, 1.0),
             xytext=((np.pi/2)+1.0, .8),
             weight="bold",
             color="b",
             arrowprops
             =dict(arrowstyle="->", connectionstyle="arc3", color="b"))

plt.show()
```

### 1.3.9 函数 text()——添加图形内容细节的无指向型注释文本

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3.2, 3.2, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend()

# plt.text(x：注释文本内容所在位置的横坐标, y：注释文本内容所在位置的纵坐标, 
# string：注释文本内容, weight：注释文本内容的粗细风格, color：注释文本内容的字体颜色)
plt.text(0.1, 0.09, "y=sin(x)", weight="bold", color="b")

plt.show()
```

### 1.3.10 函数 title()——添加图形内容的标题

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3.2, 3.2, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend()

plt.title("y=sin(x)")

plt.show()
```

### 1.3.11 函数 legend()——标示不同图形的文本标签图例

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3.2, 3.2, 100)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")
plt.legend(loc="lower left")

plt.show()
```

## 1.4 函数组合应用

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm as cm

# define data set
x = np.linspace(0.5, 3.5, 100)
y = np.sin(x)
y1 = np.random.randn(100)

# plot figure
plt.plot(x, y, ls="--", lw=2, label="plot figure")

# scatter figure
plt.scatter(x, y1, c="0.25", label="scatter figure")

# some clean up(remove chartjunk)
# turn the top spine and the right spine off
for spine in plt.gca().spines.keys():
    if spine == "top" or spine == "right":
        plt.gca().spines[spine].set_color("none")

# turn bottom tick for x-axis on
plt.gca().xaxis.set_ticks_position("bottom")
# set tick_line position of bottom

# leave left ticks for y-axis on
plt.gca().yaxis.set_ticks_position("left")
# set tick_line position of left

plt.xlim(0.0, 4.0)
plt.ylim(-3.0, 3.0)

# set axes labels
plt.xlabel("x_axis")
plt.ylabel("y_axis")

# set x, y axis grid
plt.grid(True, ls=":", color="r")

# add a horizontal line across the axis
plt.axhline(y=0.0, c="r", ls="--", lw=2)

# add a vertical span across the axis
plt.axvspan(xmin=1.0, xmax=2.0, facecolor="y", alpha=.3)

# set annotating info
plt.annotate("maximum", xy=(np.pi/2, 1.0),
             xytext=((np.pi/2)+0.15, 1.5), weight="bold", color="r",
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="r"))
plt.annotate("spines", xy=(0.75, -3),
             xytext=(0.35, -2.25), weight="bold", color="b",
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="b"))
plt.annotate("", xy=(0, -2.78),
             xytext=(0.4, -2.32),
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="b"))
plt.annotate("", xy=(3.5, -2.98),
             xytext=(3.6, -2.70),
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="b"))

plt.text(3.6, -2.70, "'|' is tickline", weight="bold", color="b")
plt.text(3.6, -2.95, "3.5 is ticklable", weight="bold", color="b")

# set title
plt.title("structure of matplotlib")

# set legend
plt.legend(loc="upper right")

plt.show()
```
