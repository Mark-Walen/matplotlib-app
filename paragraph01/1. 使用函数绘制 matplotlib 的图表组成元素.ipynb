```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.05, 10, 1000)
y = np.random.rand(1000)

plt.scatter(x, y, label="scatter figure")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3.2, 9.5, 1000)
y = np.cos(x)

plt.scatter(x, y, ls="-", lw=2, label="plot figure")
plt.legend()
plt.show()
```
